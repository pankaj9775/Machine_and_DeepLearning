```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn import ensemble
import tensorflow as tf
from sklearn import preprocessing

tf.logging.set_verbosity(tf.logging.INFO)
from sklearn import decomposition
```

```python
labels = pd.read_table('./data/Human activity/activity_labels.txt', sep=' ', header=None, names=('ID','Activity'))
feature_names = pd.read_table('./data/Human activity/features.txt', sep=' ', header=None, names=('ID', 'Sensor'))
```

```python
feature_names = np.array(feature_names.Sensor)
labels = np.array(labels.Activity)
```

```python
X_train = pd.read_table('./data/Human activity/final_X_train.txt', sep=',', header=None)
X_test = pd.read_table('./data/Human activity/final_X_test.txt', sep=',', header=None)
Y_train = pd.read_table('./data/Human activity/final_y_train.txt', sep=',', header=None)
Y_test = pd.read_table('./data/Human activity/final_y_test.txt', sep=',', header=None)

print ("Total number of features in dataset:", X_train.shape[1])
print ("Total instances in training dataset:", X_train.shape[0])
print ("Total instances in test dataset:", X_test.shape[0])
```

```python
X_train.columns = feature_names
X_test.columns = feature_names
```

```python
pca = decomposition.PCA(n_components=70)
Xtrain = pca.fit_transform(X_train)
Xtest = pca.transform(X_test)
```

```python
#train = [Xtrain, Y_train] 
train = pd.DataFrame(Xtrain)
train['target'] = Y_train

#test = [Xtest, Y_test] 
test = pd.DataFrame(Xtest)
test['target'] = Y_test
```

```python
coloumns = (train.columns)
feature_new = test.ix[:100,:-1]
feature_new = feature_new.columns
```

```python
feature_new = list(feature_new)
coloumns = list(coloumns)
```

```python
print(len(feature_new))
print(len(coloumns))
```

```python
training_set = train
test_set = test
```

```python
print(training_set.shape)
print(test_set.shape)
```

```python
lb = preprocessing.LabelEncoder()
f1 = lb.fit_transform(coloumns)
f2 = lb.transform(feature_new)
f1 = f1.astype('str')
f2 = f2.astype('str')
f1 = list(f1)
f2= list(f2)
```

```python
f1 = ["a" + f1 for f1 in f1]
f2 = ["a" + f2 for f2 in f2]
```

```python
training_set.columns = f1
test_set.columns = f1
```

```python
print(training_set.shape)
print(test_set.shape)
```

```python
del train
del test
```

```python
COLUMNS = f1
FEATURES = f2
LABEL = 'a70'
```

```python
training_set.a70 = training_set.a70.replace(to_replace=[1,2,3,4,5,6],value=[0,1,2,3,4,5])
test_set.a70 = test_set.a70.replace(to_replace=[1,2,3,4,5,6],value=[0,1,2,3,4,5])
```

```python
def input_fn(data_set):
  feature_cols = {k: tf.constant(data_set[k].values) for k in FEATURES}
  labels = tf.constant(data_set[LABEL].values)
  return feature_cols, labels

feature_cols = [tf.contrib.layers.real_valued_column(k) for k in FEATURES]

classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_cols,hidden_units=[20,20],n_classes=6,
                                            optimizer=tf.train.AdamOptimizer(learning_rate=0.01))
```

```python
classifier.fit(input_fn=lambda: input_fn(training_set),steps=20000)
```

```python
ev = classifier.evaluate(input_fn=lambda: input_fn(test_set), steps=50)
loss_score = ev["loss"]
print("Loss: {0:f}".format(loss_score))
```
